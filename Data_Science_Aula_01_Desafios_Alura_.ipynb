```python
dados.groupby('Bairro')['Metragem'].mean()
```

```python
media_bairro = dados.groupby('Bairro')['Metragem'].mean()
media_bairro_ordenada = media_bairro.sort_values(ascending=False)
media_bairro_ordenada.head(10)
```

```python
media_bairro_ordenada.plot(title ="Gráfico Média Metragem X Bairro - 5 Maiores")
media_bairro_ordenada.plot(ylabel ="Média Metragem")
media_bairro_ordenada.plot(xlabel ="Bairros")
media_bairro_ordenada.head(5).plot.bar()
```

Nos baseando no Gráfico e nas informações acima (pegando o exemplo da solicitação do Paulo - Média da Metragem por Bairro e ordenando os dados - é possível identificar que o Bairro Sítio Represa possuí maior Média de Metragem. 
Porém, observando mais a fundo os dados, pode-se perceber que só há 1 residência neste local e com grande metragem. Temos então uma discrepância de dados neste caso.

```python
media_bairro_quartos = dados.groupby('Bairro')['Quartos'].mean()
media_bairro_ordenada_quartos = media_bairro_quartos.sort_values(ascending=False)

media_bairro_ordenada_quartos.plot(title ="Gráfico Média Quartos X Bairro - 5 Maiores")

media_bairro_ordenada_quartos.head(5).plot.pie()
```

```python
media_bairro_vagas = dados.groupby('Bairro')['Vagas'].mean()
media_bairro_ordenada_vagas = media_bairro_vagas.sort_values(ascending=False)

media_bairro_ordenada_vagas.plot(title ="Gráfico Média Vagas X Bairro - 5 Maiores")

media_bairro_ordenada_vagas.head(5).plot.pie()
```

```python
media_bairro_banheiro = dados.groupby('Bairro')['Banheiros'].mean()
media_bairro_ordenada_banheiro = media_bairro_banheiro.sort_values(ascending=False)
media_bairro_ordenada_banheiro.plot(title ="Gráfico Média Banheiros X Bairro - 5 Maiores")
media_bairro_ordenada_banheiro.head(5).plot.pie()
```

```python
media_bairro_metragem = dados.groupby('Bairro')['Metragem'].mean()
media_bairro_ordenada_metragem = media_bairro_metragem.sort_values(ascending=False)
media_bairro_ordenada_metragem.plot(title ="Gráfico Média Metragem X Bairro - 5 Maiores")
media_bairro_ordenada_metragem.head(5).plot.pie()
```

Analisando acima os gráficos de pizza das Médias das cinco maiores (Metragens, Quartos, Banheiros, Vagas) é possível verficar que Jardin Camélias - sendo um dos bairros com maior m² também comporta a maior média de Banheiros, Vagas - Mas não de quartos.
